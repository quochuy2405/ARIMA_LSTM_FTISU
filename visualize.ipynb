```python
import pandas as pd
import numpy as np
```

```python
df_real = pd.read_csv('./data/clean/btc.csv')
df_fake = pd.read_csv('./final_pred_arima_lstm.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np

real_price = df_real['close'].values.reshape(-1, 1)
predict_price = df_fake['close'].values.reshape(-1, 1)

plt.figure(figsize=(100, 30))
plt.plot(predict_price, color="blue", label="Predicted BTC Closing Prices")
plt.plot(real_price, color="red", label="Real BTC Closing Prices")
plt.title("BTC Prices")
plt.xlabel("Time (day)")
plt.ylabel("Stock Prices")
plt.ylim(bottom=0)
plt.legend()
plt.show()
```
